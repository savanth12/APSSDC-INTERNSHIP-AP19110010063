## TASK - 1: 
* Take one dataset and plot all graphs.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
d = pd.read_csv("https://raw.githubusercontent.com/LavanyaPolamarasetty/Datasets/master/Classification/Social_Network_Ads.csv")
d
```

```python
d.head()
```

### Line plot:

```python
plt.plot(d["Age"])
plt.title("Line Plot")
plt.xlabel("Age")
plt.show()
```

### Bar graph:

```python
plt.bar(d["Gender"],d["EstimatedSalary"])
plt.title("Bar Graph")
plt.xlabel("Gender")
plt.ylabel("Estimated salary")
plt.show()
```

### Scatter Plot:

```python
plt.scatter(d["Age"],d["EstimatedSalary"])
plt.title("Scatter Plot")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.show()
```

### Histogram:

```python
plt.hist(d["Age"])
plt.title("Histogram")
plt.xlabel("Age")
plt.ylabel("Frequency count of age")
```

### Pie Chart:

```python
plt.pie(d["EstimatedSalary"][:10],labels = d["Age"][:10], autopct = "% 1f%%")
plt.show()
```

### Box Plot:

```python
plt.boxplot(d["EstimatedSalary"])
plt.title("Box Plot")
plt.xlabel("Estimated Salary")
plt.show()
```

### Subplot:

```python
fig, area = plt.subplots(2, 2)
area[0][0].plot(d['Age'])
area[0][1].plot(d['Gender'])
area[1][0].plot(d['EstimatedSalary'])
area[1][1].plot(d['Purchased'])
plt.show()
```

```python
plt.subplot(1, 2, 2)
plt.plot(d['EstimatedSalary'])
plt.subplot(1, 2, 1)
plt.plot(d['Gender'])
plt.show()
```
