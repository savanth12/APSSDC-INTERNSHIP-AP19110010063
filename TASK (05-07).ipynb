### TASK 1 : 
#### Take kernel = 'rbf' for the below dataset and check the accuracy

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/lung_cancer_examples.csv')
df
```

```python
df.isnull().sum()
```

```python
df.duplicated()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

model = SVC(kernel = 'rbf')
```

```python
x = df[df.columns[2:-1]]
y = df['Result']
```

```python
model.fit(x,y)
```

```python
predict = model.predict(x)
predict
```

```python
print(accuracy_score(y,predict))
```

### TASK 2 :
#### Take any dataset from Kaggle. Apply SVC model with all values of kernel attribute

```python
d = pd.read_csv('https://gist.githubusercontent.com/omairaasim/1437d57bc945ab74729eaa30fd022fbf/raw/d7e49cce8accecbbde2d2c8135f608e96fd8502e/project_10_dataset_sample.csv')
d
```

```python
x = d[d.columns[1:3]]
y = d["Purchased Iphone"]
```

### Kernel = 'linear'

```python
model = SVC(kernel = 'linear')
model.fit(x,y)
```

```python
predict = model.predict(x)
predict
```

```python
print(accuracy_score(y,predict))
```

### Kernel = 'poly'

```python
model1 = SVC(kernel = 'poly')
model1.fit(x,y)
```

```python
predict1 = model1.predict(x)
predict1
```

```python
print(accuracy_score(y,predict1))
```

### Kernel = 'rbf'

```python
model2 = SVC(kernel = 'rbf')
model2.fit(x,y)
```

```python
predict2 = model2.predict(x)
predict2
```

```python
print(accuracy_score(y,predict2))
```
