## Task - 1
* Replace the null values 'Age' column with ffill(), and values in 'Cabin' and 'Embarked' columns with the most occured value in the column

```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv("https://raw.githubusercontent.com/nagamounika5/Datasets/master/titanic.csv")
titanic
```

```python
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
titanic['Age'].fillna(method = "ffill", inplace = True)
titanic
```

```python
titanic['Cabin'].fillna(titanic['Cabin'].mode().iloc[0], inplace = True)
titanic['Embarked'].fillna(titanic['Embarked'].mode().iloc[0], inplace = True)
titanic
```

```python
titanic.isnull().sum()
```

## Task - 2
* Convert 'Cabin' and 'Embarked' columns into numerical columns

```python
from sklearn.preprocessing import LabelEncoder
lab = LabelEncoder()
titanic['Cabin'] = lab.fit_transform(titanic['Cabin'])
titanic['Embarked'] = lab.fit_transform(titanic['Embarked'])
titanic['Name'] = lab.fit_transform(titanic['Name'])
titanic['Sex'] = lab.fit_transform(titanic['Sex'])
titanic['Ticket'] = lab.fit_transform(titanic['Ticket'])
titanic
```
